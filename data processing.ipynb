```python
import pandas as pd
df = pd.read_csv('heart.csv')
df.head()
df = df[df.RestingBP != 0]
df = df[df.Cholesterol != 0]
df.Oldpeak=df.Oldpeak*10
#making bins
agegrp= pd.cut(df.Age,bins=[20,29,39,49,59,69,79],labels=["20s","30s","40s","50s","60s","70s"])
df.insert(1,"Age_Group",agegrp)
rbp= pd.cut(df.RestingBP,bins=[0,119,129,139,1000],labels=["Normal","Elevated","High","Extreme"])
df.insert(5,"Blood_Pressure_Level",rbp)
op= pd.cut(df.Oldpeak,bins=[-10,20,1000],labels=["Normal","Risk"])
df.insert(12,"Oldpeak_Level",op)
mHR= pd.cut(df.MaxHR,bins=[60,79,99,119,139,159,179,199,219],labels=["60-79","80-99","100-119","120-139","140-159","160-179","180-199","200-220"])
df.insert(10,"Max_Heart_Rate",mHR)
ch= pd.cut(df.Cholesterol,bins=[0,199,240,1000],labels=["<200","200-240",">240"])
df.insert(7,"Cholestrol_levels",ch)

df.describe()
df=df[df.Sex=="M"]
df
```
